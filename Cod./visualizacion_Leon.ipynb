```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import os
from os import listdir
from os.path import isfile, join
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
```

```python
df=pd.read_csv('/workspaces/Entrega1/data/chocolate_bars_UTF8v5 - chocolate_bars_UTF8v5.csv')
df
```

```python
df= df[df['rating'].notna()]
df.insert(17,'contador',1)
df.insert(18,'suma acum',df['contador'].cumsum())
df=df.sort_values('rating',ascending= True)
df
```

```python
df[df['rating']=='NaN']
```

```python
cp=pd.DataFrame
cp=df.groupby('id')['cocoa_percent','rating'].head(2)
cp=cp.sort_values('cocoa_percent',ascending= True)
cp

```

```python
x=cp['cocoa_percent']
y=cp['rating'] 
fig, ax = plt.subplots(figsize=(10,8))
ax.plot(x,y,label='Rating')
ax.set_xlabel('cocoa_percent')
ax.set_ylabel('rating')
ax.set_title('porcentaje de cacao v/s puntaje')
ax.legend()

```

## De la grafica podemos inferir que el porcentaje de cacao, no es necesariamente direcctamente proporcional, en el total de su rango, pero vale la pena estudiar a mayor detalle el rango 65%-75%
